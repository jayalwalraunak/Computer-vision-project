```python
import cv2
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
cap = cv2.VideoCapture(0)

while True:
    succes,frame = cap.read()
    if not succes:
        break
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    face = face_detector.detectMultiScale(gray,1.1,5)
    for(x,y,w,h) in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(217,216,215),2)
        cv2.imshow("face detection",frame)
    if cv2.waitKey(1) == ord("q"):
       break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture(0)
mode = "color"
while True:
    success,frame = cap.read()
    if not success:
        break
    if mode=="gray":
        frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    elif mode == "red":
        frame[:,:,0] = 0
        frame[:,:,1] = 0
    cv2.imshow("webcam filter",frame)
    key = cv2.waitKey(1)
    if key == ord("g"):
        mode="gray"
    elif key == ord("r"):
        mode="red"
    elif key == ord("c"):
        mode="color"
    elif key == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture(0)
face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

while True:
    succes,frame = cap.read()
    if not succes:
        break
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    face = face_detector.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30,30),
)
    for(x,y,w,h) in face:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    count_text = f"Faces :{len(face)}"
    cv2.rectangle(frame,(0,0),(200,50),(0,255,0),-1)
    cv2.putText(
        frame,
        count_text,
        (10,35),
        cv2.FONT_HERSHEY_SIMPLEX,
        1,
        (0,0,0),
        2
    )
    cv2.imshow("faces Counter",frame)
    if cv2.waitKey(1) &0xff== ord("q"):
       break
print("final face count",len(face))

cap.release()
cv2.destroyAllWindows()
```
