```python
#resize image
image = cv2.imread("WhatsApp Image 2026-06-01 at 5.14.37 PM (1).jpeg")
image_1 = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
height ,width = image.shape[:2]
```

```python
center = (width//2 ,height//2)
```

```python
roation_matrix = cv2.getRotationMatrix2D(center,45,1)
```

```python
rotated = cv2.warpAffine(image,roation_matrix,(width,height))
                        
```

```python
rotated_rgb = cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB)
```

```python
fig ,axes = plt.subplots(1,2,figsize=(10,5))
axes[0].imshow(image_1)
axes[0].set_title("original image")
axes[0].axis("off")

axes[1].imshow(rotated)
axes[1].set_title("rotated image")
axes[1].axis("off")

plt.show()
```
